```python
def add_noise(noise_free, grad, snr, rician=True):
    """
    Add Gaussian noise or Rician noise to a signal.

    Returns:
        numpy.ndarray: A signal with added Gaussian/Rician noise.
    """

    mean_b0 = np.mean(noise_free[..., np.where(grad[...,-1] == 0)[0]])

    std_noise = mean_b0 / snr

    if not rician:
        noise = np.random.normal(0, std_noise, size=noise_free.shape)
        noisy_dwi = noise_free + noise
    else:
        noise1 = np.random.normal(0, std_noise, size=noise_free.shape)
        noise2 = np.random.normal(0, std_noise, size=noise_free.shape)
        noisy_dwi = np.sqrt((noise_free + noise1)**2 + noise2**2)

    return noisy_dwi

def run_msst_csd(dwi, rf, grad):
    temp_dir = '/LOCALDATA/sleyse4/temp_sim'
    os.makedirs(temp_dir)
    save_mrtrix(os.path.join(temp_dir, 'temp_dwi.mif'), Image(dwi.reshape((-1, 1, 1, dwi.shape[-1])), grad=grad, comments=[]))
    utils.save_vector(os.path.join(temp_dir, 'wm.txt'), rf)
    utils.save_vector(os.path.join(temp_dir, 'csf.txt'), np.zeros_like(rf))
    utils.save_vector(os.path.join(temp_dir, 'gm.txt'), np.zeros_like(rf))
    csd_cmd = f'dwi2fod csd {os.path.join(temp_dir, "temp_dwi.mif")} {os.path.join(temp_dir, "wm.txt")} {os.path.join(temp_dir, "temp_odf.mif")} -force'
    csd_cmd = f'dwi2fod msmt_csd {os.path.join(temp_dir, "temp_dwi.mif")} {os.path.join(temp_dir, "wm.txt")} {os.path.join(temp_dir, "wm.mif")} {os.path.join(temp_dir, "csf.txt")} {os.path.join(temp_dir, "csf.mif")} {os.path.join(temp_dir, "gm.txt")} {os.path.join(temp_dir, "gm.mif")} -force'
    subprocess.run(csd_cmd.split(' '))
    est_odf = load_mrtrix(os.path.join(temp_dir, "wm.mif")).data

    shutil.rmtree(temp_dir)

    return np.squeeze(est_odf)

def run_simulation(sample_odf, rf, grad, snr, num_its):

    odf = np.repeat(sample_odf[np.newaxis,...], num_its, axis=0)

    noise_free_signal = sh.sphconv(rf, odf)
    noise_free_dwi = sh.calcdwi(noise_free_signal, grad)
    noisy_dwi = add_noise(noise_free_dwi, grad, snr, rician=True)

    mask = np.ones(noisy_dwi.shape[:-1], dtype=bool)
    res_dict = blind_deconvolution(noisy_dwi, mask, grad, range_ad, range_rd, reg, lmax=8, Q=Q, cores=50)

    est_odf = res_dict['odf']
    est_rf = res_dict['response']

    csd_odf = run_msst_csd(noisy_dwi, rf, grad)

    return np.squeeze(est_odf), np.squeeze(est_rf), csd_odf

def get_odf_distances(est_odf, gt_odf):
    full_gt_odf = np.repeat(gt_odf[np.newaxis, ...], est_odf.shape[0], axis=0).reshape(est_odf.shape)
    return sh.angularCorrelation(est_odf, full_gt_odf)


def fill_kernels(kernels, ax, color):
    # Calculate the maximum degree of spherical harmonics
    lmax = (kernels[0].shape[-1]-1)*2

    # Get the number of shells in the kernel
    num_shells = len(kernels[0])

    # Generate a modified spherical harmonics basis
    Q_k = sh.modshbasis(lmax, np.linspace(0, 2*np.pi, 400), np.zeros(400))

    # Initialize indices for spherical harmonics
    idx = []
    end = 0
    for l in range(0, lmax + 1, 2):
        idx.append(end + (2 * l + 1) // 2)
        end += 2 * l + 1

    # Select the required columns from the basis
    Q_k = Q_k[:,idx]

    # Convert kernels to polar coordinates
    kernels_polar = [kernel @ Q_k.T for kernel in kernels]
    kernels_polar = np.asarray(kernels_polar)

    # Define angles for plotting
    angles = np.linspace(0, 2*np.pi, 400)
    # Plot each shell
    for i in range(num_shells):
        minima = np.squeeze(np.amin(kernels_polar[:,i,:], axis=0))
        maxima = np.squeeze(np.amax(kernels_polar[:,i,:], axis=0))



        # Fill between minima and maxima for each shell
        ax.fill_between(angles - np.pi / 2, minima, maxima, alpha=.2, color=color, edgecolor=(0,0,0,0), linewidth=0)

        # Set xticks if axis is provided
        xlocs = np.arange(0, 2 * np.pi, np.pi / 4)
        ax.set_xticks(xlocs, [''] * len(xlocs))
```

```python
from matplotlib.pyplot import setp

# function for setting the colors of the box plots pairs
def setBoxColors(bp):
    setp(bp['boxes'][0], color='blue')
    setp(bp['caps'][0], color='blue')
    setp(bp['caps'][1], color='blue')
    setp(bp['whiskers'][0], color='blue')
    setp(bp['whiskers'][1], color='blue')
    setp(bp['medians'][0], color='blue')

    setp(bp['boxes'][1], color='orange')
    setp(bp['caps'][2], color='orange')
    setp(bp['caps'][3], color='orange')
    setp(bp['whiskers'][2], color='orange')
    setp(bp['whiskers'][3], color='orange')
    setp(bp['medians'][1], color='orange')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(5*1.5, 3*1.5))
snrs = [10, 20, 50, np.inf]
snr_strings = ['10', '20', '50', r'$\infty$']
for (fiber_config, ax) in zip([single_fiber, crossing_60], axs.T):
    for i, snr in enumerate(snrs):
        positions = (1+3*i, 2+3*i)
        widths = .6
        est_odf, est_rf, csd_odf = run_simulation(fiber_config, sane_rf, grad, snr, 500)
        odf_distances_lore = get_odf_distances(est_odf, fiber_config)
        odf_distances_csd = get_odf_distances(csd_odf, fiber_config)
        # data_odf = full_df.loc[f'{fiber_config}_{snr}', ['ODF Distance BD', 'ODF Distance CSD']]
        bp = ax.boxplot([odf_distances_lore, odf_distances_csd], positions=positions, widths=widths, showfliers=False)
        setBoxColors(bp)
    ax.set_xticks([1.5+3*i for i in range(len(snrs))])
    print(snr_strings)
    ax.set_xticklabels(snr_strings)
    ax.set_xlabel('SNR')
    ax.set_ylim(.65, 1)
    if ax != axs.T[0]:
        ax.set_yticklabels([])

axs[0].set_ylabel('ODF Angular Correlation Coefficient')
for (title, ax) in zip(['Single Fibre', r'Crossing Fibre $(60\degree)$'], axs.T):
    ax.set_title(title)

# draw temporary red and blue lines and use them to create a legend
hB, = axs[-1].plot(5, 1,'-', color='blue', label='hB')
hR, = axs[-1].plot(5, 1,'-', color='orange', label='hR')
patches, _ = axs[-1].get_legend_handles_labels()
fig.legend(patches, ['LoRE-SD', 'MSMT-CSD'], loc='right')
hB.set_visible(False)
hR.set_visible(False)
plt.show()
```
