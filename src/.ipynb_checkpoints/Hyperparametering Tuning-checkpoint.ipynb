```python
import bd_gaussians
import BlindDeconvolution as bd
import utils
import phantom_generation as ph
import SphericalHarmonics as sh
import Rank1
import contrasts

from mrtrix.io import load_mrtrix, save_mrtrix
from mrtrix.io.image import Image

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import subprocess
import shutil
```

# Constants

```python
PREPROC_DIR = '/usr/local/micapollo01/MIC/DATA/STAFF/sleyse4/u0152170/Blind_Deconvolution/Preprocessing/philips'

lmax = 8

# Rounding and extracting b-values from the gradient table.
grad = load_mrtrix(os.path.join(PREPROC_DIR, 'normed.mif')).grad
grad[...,-1] = np.round(grad[...,-1], -2)
bvals = np.unique(grad[...,-1])

cmd_dirs = f'dirgen {300} dirs_full.txt -force'
subprocess.run(cmd_dirs.split(' '))
dirs = utils.load_vector('dirs_full.txt')
subprocess.run('rm dirs_full.txt'.split(' '))

Q = sh.modshbasis(lmax, dirs[:,0], dirs[:,1])
```

```python
AD = np.linspace(0, 4e-3, 10)
RD = np.linspace(0, 4e-3, 10)
```

```python
def add_noise(noise_free, grad, snr):
    """
    Add Gaussian noise or Rician noise to a signal.

    Returns:
        numpy.ndarray: A signal with added Gaussian/Rician noise.
    """
    mean_b0 = np.mean(noise_free[..., np.where(grad[...,-1] == 0)[0]])

    std_noise = mean_b0 / snr
    noise1 = np.random.normal(0, std_noise, size=noise_free.shape)
    noise2 = np.random.normal(0, std_noise, size=noise_free.shape)
    noisy_dwi = np.sqrt((noise_free + noise1)**2 + noise2**2)

    return noisy_dwi


def run_msst_csd(dwi, rf, grad):
    temp_dir = '/LOCALDATA/sleyse4/temp_sim'
    os.makedirs(temp_dir)
    save_mrtrix(os.path.join(temp_dir, 'temp_dwi.mif'), Image(dwi.reshape((-1, 1, 1, dwi.shape[-1])), grad=grad, comments=[]))
    utils.save_vector(os.path.join(temp_dir, 'wm.txt'), rf)
    utils.save_vector(os.path.join(temp_dir, 'csf.txt'), np.zeros_like(rf))
    utils.save_vector(os.path.join(temp_dir, 'gm.txt'), np.zeros_like(rf))
    csd_cmd = f'dwi2fod msmt_csd {os.path.join(temp_dir, "temp_dwi.mif")} {os.path.join(temp_dir, "wm.txt")} {os.path.join(temp_dir, "wm.mif")} {os.path.join(temp_dir, "csf.txt")} {os.path.join(temp_dir, "csf.mif")} {os.path.join(temp_dir, "gm.txt")} {os.path.join(temp_dir, "gm.mif")} -force'
    subprocess.run(csd_cmd.split(' '))
    est_odf = load_mrtrix(os.path.join(temp_dir, "wm.mif")).data

    shutil.rmtree(temp_dir)

    return np.squeeze(est_odf)
```

# Voxel Simulation

```python
odf_sim = .5*ph.getOrientedApodizedDirac(0, 0, lmax) + .5*ph.getOrientedApodizedDirac(45, 0, lmax)

stick = bd.zhgaussian(bvals, 2.2e-3, 0)
zeppelin = bd.zhgaussian(bvals, 2e-3, .7e-3)
water = bd.zhgaussian(bvals, 3e-3, 3e-3)

rf_sim = .9*(.5*stick + .5*zeppelin) + .1*water
# rf_sim *= 1000/np.sqrt(4*np.pi)
rf_sim *= 1000

sh_sim = sh.sphconv(rf_sim, odf_sim)
dwi_sim = sh.calcdwi(sh_sim, grad)
```

```python
fig, axs = plt.subplots(1, 2, subplot_kw={'projection': 'polar'})

bd.plot_odf(odf_sim, ax=axs[0])
axs[0].set_title('Simulated ODF')

bd.plot_wmr(rf_sim, bvals=bvals, ax=axs[1])
axs[1].set_title('Simulated Response Function')
plt.show()
```

```python
reps = 500

noise_free = np.repeat(dwi_sim, reps, axis=0)

dwi_noise_10 = add_noise(noise_free, grad, 10)
dwi_noise_20 = add_noise(noise_free, grad, 20)
dwi_noise_50 = add_noise(noise_free, grad, 50)

dwi = np.asarray([dwi_noise_10, dwi_noise_20, dwi_noise_50, noise_free])
mask = np.ones(dwi.shape[:-1], dtype=bool)

noise_10 = Rank1.residual(dwi_noise_10, noise_free, np.ones(dwi_noise_10.shape[:-1], dtype=bool))
noise_20 = Rank1.residual(dwi_noise_20, noise_free, np.ones(dwi_noise_20.shape[:-1], dtype=bool))
noise_50 = Rank1.residual(dwi_noise_50, noise_free, np.ones(dwi_noise_50.shape[:-1], dtype=bool))
no_noise = np.zeros_like(noise_10)
noise = np.asarray([noise_10, noise_20, noise_50, no_noise])
```

```python
sh.pretty_print_gradient_table(grad)
```

```python
odf_est_msmt_csd = run_msst_csd(dwi, rf_sim, grad).reshape((4, reps, -1))
acc_msmt_csd = sh.angularCorrelation(odf_sim, odf_est_msmt_csd)
res_msmt_csd = Rank1.residual(sh.calcdwi(sh.sphconv(rf_sim, odf_est_msmt_csd), grad), dwi, mask)
```

# Hyperparameter Tuning: Regularisation parameter $\lambda$

```python
log_test = np.logspace(-7, 0, 8)
full_acc_values = []
full_residuals = []
for reg in log_test:
    result = bd_gaussians.blind_deconvolution(dwi, mask, grad, AD, RD, reg=reg, Q=Q)
    acc = sh.angularCorrelation(odf_sim, result['odf'])
    residual = result['residual']
    full_acc_values.append(acc)
    full_residuals.append(residual)

full_acc_values = np.asarray(full_acc_values)
full_residuals = np.asarray(full_residuals)
```

```python
import PIL

def trim(im, only_horizontal=False):
    im = im.convert('RGB')
    bg = PIL.Image.new(im.mode, im.size, im.getpixel((0, 0)))
    diff = PIL.ImageChops.difference(bg, im)
    bbox = diff.getbbox()
    if bbox:
        if only_horizontal:
            bbox = (bbox[0], 0, bbox[2], im.height-1)
        cropped = im.crop(bbox)
        return cropped

os.makedirs('cropped_images', exist_ok=True)
for filename in os.listdir('images'):
    if filename.endswith('.png') or filename.endswith('jpg'):
        im = PIL.Image.open(f'images/{filename}').convert('RGB')
        im = trim(im, only_horizontal=True)
        newFilePath = f'cropped_images/{filename}'
        im.save(newFilePath)
```

```python
dataset = 'philips'
sub = dataset.split('/')[-1]

brain_slice = 34

# Data directory
PREPROC_DIR = f'/usr/local/micapollo01/MIC/DATA/STAFF/sleyse4/u0152170/Blind_Deconvolution/Preprocessing/{sub}'
DATA_DIR = f'/usr/local/micapollo01/MIC/DATA/STAFF/sleyse4/u0152170/DATA/{dataset}/'
SAVE_DIR = f'/usr/local/micapollo01/MIC/DATA/STAFF/sleyse4/u0152170/Blind_Deconvolution/temp/{sub}/grid_size'
MTCSD_DIR = f'/usr/local/micapollo01/MIC/DATA/STAFF/sleyse4/u0152170/Blind_Deconvolution/Preprocessing/{sub}/MT-CSD'

os.makedirs(SAVE_DIR, exist_ok=True)
```

```python
def full_brain_csd(input_dwi_path, tissue_segmentation_path, gt_odf, gt_dwi, mask):
    temp_dir = '/LOCALDATA/sleyse4/temp'
    os.makedirs(temp_dir, exist_ok=True)
    subprocess.run(['dwi2response', 'msmt_5tt', input_dwi_path, tissue_segmentation_path, f'{temp_dir}/wm.txt', f'{temp_dir}/gm.txt', f'{temp_dir}/csf.txt', '-force', '-nthreads', '50'])

    data = load_mrtrix(input_dwi_path)
    save_mrtrix(f'{temp_dir}/temp_dwi.mif', Image(data=data.data, vox=data.vox, grad=data.grad, comments=[]))

    subprocess.run(['dwi2fod', 'msmt_csd', f'{temp_dir}/temp_dwi.mif', f'{temp_dir}/wm.txt', f'{temp_dir}/wm.mif', f'{temp_dir}/gm.txt', f'{temp_dir}/gm.mif', f'{temp_dir}/csf.txt', f'{temp_dir}/csf.mif', '-predicted_signal', f'{temp_dir}/prediction.mif', '-force', '-nthreads', '20'])

    wm = load_mrtrix(f'{temp_dir}/wm.mif').data
    acc = sh.angularCorrelation(wm, gt_odf)
    rec = load_mrtrix(f'{temp_dir}/prediction.mif').data

    subprocess.run(['rm', '-r', temp_dir])

    return acc, rec

def RSS(target, pred, mask, subtract_mean=False):
    residual = target - pred
    if subtract_mean:
        residual -= np.mean(residual, axis=-1)

    out = np.sum(residual ** 2, axis=-1)

    out[~mask] = 0
    return out


def aic(k, target, pred, mask, subtract_mean=False):
    n = target.shape[-1]
    sigma_sq = RSS(target, pred, mask, subtract_mean) / n
    out = 2 * k + n * np.log(sigma_sq)
    out[~mask] = 0
    return out

def mask2edge(binary_mask):
    if len(binary_mask.shape) == 2:
        gx, gy, = np.gradient(binary_mask.astype(float))
        edge = gy * gy + gx * gx
    elif len(binary_mask.shape) == 3:
        gx, gy, gz = np.gradient(binary_mask.astype(float))
        edge = gy * gy + gx * gx + gz * gz
    else:
        raise Exception
    edge[edge != 0] = 1
    return edge

def pad_to_shape(data, shape):
    data_shape = data.shape

    pad_x = shape[0] - data_shape[0]
    pad_left = pad_x // 2
    pad_right = pad_x - pad_left

    pad_y = shape[1] - data_shape[1]
    pad_bottom = pad_y // 2
    pad_top = pad_y - pad_bottom

    return np.pad(data, ((pad_left, pad_right), (pad_bottom, pad_top)))

def clip(img, mask):
    mask_x = np.unique(np.where(mask)[0])
    mask_y = np.unique(np.where(mask)[1])
    x_min, x_max = np.min(mask_x), np.max(mask_x)
    y_min, y_max = np.min(mask_y), np.max(mask_y)
    return img[x_min:x_max+1, y_min:y_max+1]

def overlay_lesion(lesion, mask, output_shape, ax):
    lesion_edge = clip(mask2edge(lesion), mask)
    lesion_edge = pad_to_shape(lesion_edge, output_shape)
    ax.imshow(np.rot90(lesion_edge), alpha=np.clip(np.rot90(lesion_edge), a_min=0, a_max=1), cmap='winter')
```

```python
grid_test = [3,5,7,10,15,20]

data = load_mrtrix(os.path.join(PREPROC_DIR, 'normed.mif'))
mask = (load_mrtrix(os.path.join(DATA_DIR, 'PREPROC/mask.mif')).data > .5)

bbox = np.argwhere(mask)

# Compute the bounding box

(ystart, xstart, _), (ystop, xstop, _) = bbox.min(0), bbox.max(0) + 1

dwi = data.data

grad = data.grad
bvals = np.unique(np.round(grad[...,-1], -2))

acc = np.empty((len(grid_test),) + dwi.shape[:3])

res_lore = np.empty((len(grid_test),) + dwi.shape[:3])
aic_lore = np.empty((len(grid_test),) + dwi.shape[:3])
iax = np.empty((len(grid_test),) + dwi.shape[:3])
eax = np.empty((len(grid_test),) + dwi.shape[:3])
fw = np.empty((len(grid_test),) + dwi.shape[:3])

csd_res = load_mrtrix(os.path.join(MTCSD_DIR, 'residual.mif')).data
csd_odf = load_mrtrix(os.path.join(MTCSD_DIR, 'wm.mif')).data
csd_aic = load_mrtrix(os.path.join(MTCSD_DIR, 'AIC.mif')).data

for i, num in enumerate(grid_test):
    ad_range = np.linspace(0, 4e-3, num)
    rd_range = np.linspace(0, 4e-3, num)

    intra_axonal_comp = contrasts.exponential_stick_weighting(ad_range, rd_range, weight=2)
    free_water_comp = contrasts.free_water_weighting(ad_range, rd_range)
    extra_axonal_comp = 1 - intra_axonal_comp - free_water_comp

    res = bd_gaussians.blind_deconvolution(dwi, mask, grad, ad_range, rd_range, 1e-3, lmax, Q, cores=50)

    iax_contrast = np.sum(intra_axonal_comp * res['fs'], axis=(-1, -2))
    eax_contrast = np.sum(extra_axonal_comp * res['fs'], axis=(-1, -2))
    fw_contrast = np.sum(free_water_comp * res['fs'], axis=(-1, -2))

    est_odf = res['odf']
    rec = sh.calcdwi(sh.sphconv(res['response'], res['odf']), grad)
    residual = Rank1.residual(dwi, rec, mask)

    cos_sim_odf = np.nan_to_num(sh.angularCorrelation(est_odf, csd_odf))

    n_variables = 45 + sum([i for i in range(num+1)])
    print(num, n_variables)

    aic_lore[i] = aic(n_variables, dwi, rec, mask, subtract_mean=False)

    acc[i] = cos_sim_odf
    res_lore[i] = residual
    iax[i] = iax_contrast
    eax[i] = eax_contrast
    fw[i] = fw_contrast

model_selection = aic_lore - csd_aic
```

```python
from matplotlib.colors import PowerNorm

from matplotlib.colors import LinearSegmentedColormap

aic_diff_cmap = LinearSegmentedColormap.from_list('AIC_Gradient_CMap', (
    (0.000, (0, 0, 1, .6)),
    (0.500, (1, 1, 1, 1)),
    (1.000, (1, .5, 0, .6))))

fig = plt.figure(figsize=(15,8))

gs = fig.add_gridspec(3, len(grid_test)+1, width_ratios=[1 for _ in range(len(grid_test))] + [.1])

for i, grid_size in enumerate(grid_test):
    ax_acc = fig.add_subplot(gs[0, i])
    ax_res = fig.add_subplot(gs[1, i])
    ax_sel = fig.add_subplot(gs[2, i])

    vmin_model = -np.max([np.min(model_selection[i]), np.max(model_selection[i])])
    vmax_model = -vmin_model

    data_acc = np.ma.masked_where(mask==0, acc[i])
    data_res = np.ma.masked_where(mask==0, res_lore[i])
    data_sel = np.ma.masked_where(mask==0, model_selection[i])

    im_acc = ax_acc.imshow(np.rot90(data_acc[ystart:ystop, xstart:xstop]), interpolation=None, cmap='copper', norm=PowerNorm(vmin=0, vmax=1, gamma=10))
    im_res = ax_res.imshow(np.rot90(data_res[ystart:ystop, xstart:xstop]), interpolation=None, cmap='inferno', vmin=0, vmax=80)
    im_sel = ax_sel.imshow(np.rot90(data_sel[ystart:ystop, xstart:xstop]), interpolation=None, cmap=aic_diff_cmap, vmin=vmin_model, vmax=vmax_model)


    utils.clean_axis(ax_acc)
    utils.clean_axis(ax_res)
    utils.clean_axis(ax_sel)

    ax_acc.set_title(grid_size, fontsize=14)

    if 'glioma' in DATA_DIR:
        lesion = load_mrtrix(os.path.join(DATA_DIR, 'LESION/lesion_regrid.mif')).data[ystart:ystop, xstart:xstop,brain_slice] > .5
        overlay_lesion(lesion, mask[ystart:ystop, xstart:xstop], data_acc[ystart:ystop, xstart:xstop].shape, ax_acc)
        overlay_lesion(lesion, mask[ystart:ystop, xstart:xstop], data_res[ystart:ystop, xstart:xstop].shape, ax_res)
        overlay_lesion(lesion, mask[ystart:ystop, xstart:xstop], data_sel[ystart:ystop, xstart:xstop].shape, ax_sel)

    if i == len(grid_test)-1:
        ax_cbar_acc = fig.add_subplot(gs[0, len(grid_test)])
        cbar_acc = plt.colorbar(im_acc, cax=ax_cbar_acc, ticks=[0, 0.8, 0.9, 0.95, 1])

        ax_cbar_res = fig.add_subplot(gs[1, len(grid_test)])
        cbar_res = plt.colorbar(im_res, cax=ax_cbar_res)

        ax_cbar_sel = fig.add_subplot(gs[2, len(grid_test)])
        cbar_sel = plt.colorbar(im_sel, cax=ax_cbar_sel, ticks=[vmin_model, vmax_model])
        cbar_sel.ax.set_yticklabels(['LoRE-SD', 'MSMT-CSD'])


    if i == 0:
        ax_acc.set_ylabel('ACC', fontsize=14)
        ax_res.set_ylabel(r'Residual$_{LoRE-SD}$', fontsize=14)
        ax_sel.set_ylabel('Model Selection', fontsize=14)
plt.suptitle(r'Grid Size ($n \times n$)', fontsize=18)
plt.savefig('images/grid_size_slice.png')
```

```python
log_test = np.logspace(-7, 0, 8)

data = load_mrtrix(os.path.join(PREPROC_DIR, 'normed.mif'))
mask = (load_mrtrix(os.path.join(DATA_DIR, 'PREPROC/mask.mif')).data > .5)[:, :, brain_slice:brain_slice + 1]

bbox = np.argwhere(mask)

# Compute the bounding box

(ystart, xstart, _), (ystop, xstop, _) = bbox.min(0), bbox.max(0) + 1

dwi = data.data[:, :, brain_slice:brain_slice + 1]

grad = data.grad
bvals = np.unique(np.round(grad[..., -1], -2))

acc = np.empty((len(log_test),) + dwi.shape[:3])

res_lore = np.empty((len(log_test),) + dwi.shape[:3])
iax = np.empty((len(log_test),) + dwi.shape[:3])
eax = np.empty((len(log_test),) + dwi.shape[:3])
fw = np.empty((len(log_test),) + dwi.shape[:3])

csd_res = load_mrtrix(os.path.join(MTCSD_DIR, 'residual.mif')).data[:, :, brain_slice:brain_slice + 1]
csd_odf = load_mrtrix(os.path.join(MTCSD_DIR, 'wm.mif')).data[:, :, brain_slice:brain_slice + 1]
```

```python
for i, reg in enumerate(log_test):
    ad_range = np.linspace(0, 4e-3, 10)
    rd_range = np.linspace(0, 4e-3, 10)

    res = bd_gaussians.blind_deconvolution(dwi, mask, grad, ad_range, rd_range, reg, lmax, Q, cores=50)

    est_odf = res['odf']
    rec = sh.calcdwi(sh.sphconv(res['response'], res['odf']), grad)
    residual = Rank1.residual(dwi, rec, mask)

    cos_sim_odf = np.nan_to_num(sh.angularCorrelation(est_odf, csd_odf))

    acc[i] = cos_sim_odf
    res_lore[i] = residual
```
